```python
plt.figure(figsize = (15,5))
plt.subplot(2,2,1)
plt.plot(df.index, df[0]**2)

plt.subplot(2,2,2)
plt.plot(df.index, df[1])
plt.subplot(2,2,3)
plt.plot(df.index, df[2]*100)
plt.subplot(2,2,4)
plt.plot(df.index, df[3])
plt.xlabel('x -axis')
plt.ylabel('y -axis')
plt.title('Linear Graph')
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,2)
plt.plot(df.index, df[0]**2, label = 'line-1')
plt.plot(df.index, df[3], label = 'line-2')
plt.legend() # requires label to be set

plt.subplot(1,2,1)
plt.plot(df.index, df[1])
plt.xlabel('x -axis')
plt.ylabel('y -axis')
plt.xticks(range(20))
plt.title('Linear Graph')

plt.show()
```
